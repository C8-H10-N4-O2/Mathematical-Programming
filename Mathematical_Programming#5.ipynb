```python
import numpy as np
import sympy as sp
import math
import matplotlib.pyplot as plt

sp.init_printing()
```

```python
#1-(1)

rad = np.arange(0,(2*np.pi), 0.02)
for i in rad:
    r = 1 + (2*np.sin(i))
    if r>=0:
         plt.polar(i,r,'g.')
    else:
         plt.polar(i+np.pi,-r,'g.')
plt.show()
```

```python
#1-(2)

rad = np.arange(0,(2*np.pi), 0.02)
for i in rad:
    r = 2*np.cos(i)
    if r>=0:
         plt.polar(i,r,'g.')
    else:
         plt.polar(i+np.pi,-r,'g.')
            
plt.show()
```

```python
#1-(3)
theta = np.linspace(0,2*np.pi,1000)
r = np.e**(theta)

plt.polar(theta,r)
plt.show()
```

```python
#1-(4)
theta = np.linspace(-np.pi/2,3*np.pi/2,1000)
r=1/((abs(theta-np.pi/2)-np.pi/4)**2+1) - 1/(abs(theta-np.pi/2)-1.45*np.pi)

plt.polar(theta,r)
plt.show()
```

```python
#1-(5)
theta = np.linspace(-np.pi/2,3*np.pi/2,1000)
r = np.e**(np.sin(theta))-2*np.cos(4*theta)

#Butterfly Curve

plt.polar(theta,r)
plt.show()
```
