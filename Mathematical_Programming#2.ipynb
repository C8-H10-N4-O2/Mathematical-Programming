```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp 
```

```python
#문제 1번

def factorial(a):
    if (a == 0) : return 1
    sum = a * factorial(a-1)
    
    return sum

print('예시 : ',factorial(5))
```

```python
#문제 2번

x=np.linspace(-4, 4, 1000)
y=np.linspace(-1, 1, 1000)

y1 = np.sin(x)
y2 = x
y3 = x - (x**3 / factorial(3))
y4 = x - (x**3 / factorial(3)) + (x**5 / factorial(5))
y5 = x - (x**3 / factorial(3)) + (x**5 / factorial(5)) - (x**7 / factorial(7))

#(1)
plt.xlim(-4, 4)
plt.ylim(-1, 1)

#(2)
plt.xlabel('$x$ axis')
plt.ylabel('$y$ axis')

#(3)
plt.title('Taylor series of sin $x$')

#(4)
plt.plot(x,y1,label='$y=\sin x$',color='red',linewidth=2, linestyle='-')
plt.plot(x,y2,label='$y=f_1(x)$',color='orange',linewidth=2, linestyle='--')
plt.plot(x,y3,label='$y=f_2(x)$',color='yellow',linewidth=2, linestyle=':')
plt.plot(x,y4,label='$y=f_3(x)$',color='green',linewidth=2, linestyle='-.')
plt.plot(x,y5,label='$y=f_4(x)$',color='blue',linewidth=2, linestyle=':')
plt.legend()

#(5)
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.show()
```
